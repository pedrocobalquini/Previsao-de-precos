Excluindo colunas que tem valores nulos maiores que 10% dos dados.

```python
eliminar = df.columns[(df.isnull().sum()/df.shape[0]) > 0.1]
eliminar
```

```python
df = df.drop(eliminar, axis=1)
```

Deixando apenas valores numericos.

```python
df_num = df.select_dtypes(include=['float64', 'int64'])
df_num
```

```python
df_num.isnull().sum().sort_values(ascending=False).head(10)
```

```python
df_num = df_num.fillna(-1)
```

Criação de modelo

```python
X = df_num.drop('SalePrice', axis=1)
y = df_num.SalePrice
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

Regressão Linear

```python
reg = LinearRegression().fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

Avaliar dados com o erro quadrático médio e erro médio absoluto.

```python
print(f'Erro médio absoluto: {mean_absolute_error(y_test, y_pred)}\n')
print(f'Erro quadrático médio:{mean_squared_error(y_test, y_pred)}')
```
